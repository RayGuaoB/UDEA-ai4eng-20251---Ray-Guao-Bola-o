```python
import pandas as pd
df = pd.read_csv("train.csv")
df.head()
```

# Rellenar valores null en la columna

Primero verificamos los valores null

```python
print(df.isnull().sum()) # Verificacion numerica de valores null en todo el documento

df[df["FAMI_ESTRATOVIVIENDA"].isnull()] # verificacion en tabla de valores null FAMI_ESTRATOVIVIENDA
```

Se observa que tenemos 32137 valores null para la columna FAMI_ESTRATOVIVIENDA

Rellenamos los valores null de acuerdo a la moda, ya que es un valor categorico

```python
df['FAMI_ESTRATOVIVIENDA'] = df['FAMI_ESTRATOVIVIENDA'].fillna(df['FAMI_ESTRATOVIVIENDA'].mode()[0]) # Rellenamos los valores null con la moda

df['FAMI_ESTRATOVIVIENDA'].head(50)
```

Finalmente podemos observar que en los valores 21 y 49 que anteriormente eran valores null, ahora tienen valores de acuerdo a la moda

```python
print(df.isnull().sum())
```

# Codificación etiqueta FAMI_ESTRATOVIVIENDA (One-Hot Encoding)

```python
df["FAMI_ESTRATOVIVIENDA"].unique() #Valores en columna
```

```python
df["FAMI_ESTRATOVIVIENDA_encoded"] = pd.Categorical(
    df["FAMI_ESTRATOVIVIENDA"],
    categories=["Sin Estrato", "Estrato 1", "Estrato 2", "Estrato 3", "Estrato 4", "Estrato 5", "Estrato 6"],
    ordered=True,
).codes

df[["FAMI_ESTRATOVIVIENDA", "FAMI_ESTRATOVIVIENDA_encoded"]].drop_duplicates().sort_values("FAMI_ESTRATOVIVIENDA_encoded")
```
